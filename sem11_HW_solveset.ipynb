```python
import sympy as sp
```

```python
# Выражение
```

```python
x = sp.Symbol('x')
expr = sp.sin(x)**2 - sp.cos(x)**2
expr
```

```python
# Определение корней
```

```python
expr_solve = sp.solveset(expr)
expr_solve
```

```python
# График функции
```

```python
graph = sp.plot(expr)
graph
```

```python
# Нахождение производной
```

```python
expr_diff = sp.diff(expr)
expr_diff
```

```python
# интервалы, на которых функция возрастает (f'(x) > 0)
```

```python
expr_diff_up = sp.solveset(expr_diff > 0)
expr_diff_up
```

```python
# интервалы, на которых функция убывает (f'(x) < 0)
```

```python
expr_diff_down = sp.solveset(expr_diff < 0)
expr_diff_down
```

```python
# Вычисление вершины
```

```python
expr_diff_zero = sp.solveset(expr_diff)
expr_diff_zero
```

```python
# Промежутки, на которых f > 0 (f(x) > 0)
```

```python
expr_up = sp.solveset(expr > 0)
expr_up
```

```python
# Промежутки, на которых f < 0 (f(x) < 0)
```

```python
expr_down = sp.solveset(expr < 0)
expr_down
```
